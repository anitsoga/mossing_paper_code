```python
model_params,model_XXs,model_YYs,model_etas = [np.genfromtxt(filename, delimiter=',') for filename in filenames]
```

```python
nwt = model_params.shape[0]
nsize = 6
ncontrast = 6
nN = nsize*ncontrast
nP = 2
nQ = 4
nS = 2
nT = 1
ntypes = nQ*nS*nT
couplings = np.zeros((nwt,nsize,ncontrast,ntypes,ntypes))
for iwt in range(model_params.shape[0]):
    these_params = model_params[iwt].reshape((8,nQ))
    these_XX = model_XXs[iwt].reshape((nN,nP*nS*nT))
    these_YY = model_YYs[iwt].reshape((nN,nQ*nS*nT))
    Eta = model_etas[iwt].reshape((nN,nQ*nS*nT))
    Wmx = these_params[:2,:]
    Wmy = these_params[2:6,:]
    K = these_params[6,:]
    s02 = these_params[7,:]
    T = np.array(())
    kappa = 1
    WWmx = calnet.utils.gen_Weight_k_kappa_t(Wmx,K,kappa,T,nS=nS,nT=nT)
    WWmy = calnet.utils.gen_Weight_k_kappa_t(Wmy,K,kappa,T,nS=nS,nT=nT)
    these_residuals = Eta - these_XX @ WWmx - these_YY @ WWmy
    Xi = np.zeros_like(Eta)
    tiled_s02 = np.tile(s02,nS*nT)
    rate_fprime = calnet.utils.fprime_m_miller_troyer
    phis = np.zeros((nsize,ncontrast,nQ*nS*nT))
    for istim in range(nN):
        iistim,jjstim = np.unravel_index(istim,(nsize,ncontrast))
        phis[iistim,jjstim] = rate_fprime(Eta[istim],Xi[istim]**2+tiled_s02)
        Phi = np.diag(phis[iistim,jjstim])
        couplings[iwt,iistim,jjstim] = Phi @ np.linalg.inv(np.eye(nQ*nS*nT) - WWmy @ Phi)
```

```python
basefold = '/Users/dan/Documents/notebooks/mossing-PC/shared_data/calnet_data/'
def get_weights_files(fit_lbl):
    subfold = 'weights_%s/'%fit_lbl
    plots_fold = basefold + 'fit_graphics/' + subfold
    weights_fold = basefold + 'weights/' + subfold
    weights_files = glob.glob(weights_fold+'*.npy')
    weights_files.sort()
    weights_filenames = [weights_file.split('/')[-1][:-4] for weights_file in weights_files]
    return weights_files,weights_filenames
```

```python
def read_list(Wstar_dict):
    try:
        Wmx,Wmy,Wsx,Wsy,s02,K,kappa,T,XX,XXp,Eta,Xi,h1,h2 = Wstar_dict['as_list']
        bl = np.zeros((nQ,))
        amp = np.ones((nQ*nS*nT,))
    except:
        try:
            Wmx,Wmy,Wsx,Wsy,s02,K,kappa,T,XX,XXp,Eta,Xi,h1,h2,bl = Wstar_dict['as_list']
            amp = np.ones((nQ*nS*nT,))
        except:
            Wmx,Wmy,Wsx,Wsy,s02,K,kappa,T,XX,XXp,Eta,Xi,h1,h2,bl,amp = Wstar_dict['as_list']
    return Wmx,Wmy,Wsx,Wsy,s02,K,kappa,T,XX,XXp,Eta,Xi,h1,h2,bl,amp
```

```python
ifit = -1
lcutoff = np.nanpercentile(losses[ifit],20)
good_wts = np.where(losses[ifit]<lcutoff)[0]
nwt = len(good_wts)
all_params = np.zeros((nwt,8,4))
all_XXs = np.zeros((nwt,36,4))
all_YYs = np.zeros((nwt,36,8))
all_etas = np.zeros((nwt,36,8))
all_residuals = np.zeros((nwt,36,8))
for iiwt,iwt in enumerate(good_wts):
#     these_vals = [Wmxs[ifit][iwt],Wmys[ifit][iwt],Ks[ifit][iwt],s02s[ifit][iwt]]
    these_vals = [Wmxs[ifit][iwt],Wmys[ifit][iwt],Ks[ifit][iwt],bls[ifit][iwt]]
    for ival in range(len(these_vals)):
        if len(these_vals[ival].shape)==1:
            these_vals[ival] = these_vals[ival][np.newaxis]
    these_vals = np.concatenate(these_vals,axis=0)
    all_params[iiwt] = these_vals
    
    these_vals = [XXs[ifit][iwt]]
    for ival in range(len(these_vals)):
        if len(these_vals[ival].shape)==1:
            these_vals[ival] = these_vals[ival][np.newaxis]
    these_vals = np.concatenate(these_vals,axis=0)
    all_XXs[iiwt] = these_vals.reshape((36,4))
    
    these_vals = [YYs[ifit][iwt]]
    for ival in range(len(these_vals)):
        if len(these_vals[ival].shape)==1:
            these_vals[ival] = these_vals[ival][np.newaxis]
    these_vals = np.concatenate(these_vals,axis=0)
    all_YYs[iiwt] = these_vals.reshape((36,8))
    
    these_vals = [Etas[ifit][iwt]]
    for ival in range(len(these_vals)):
        if len(these_vals[ival].shape)==1:
            these_vals[ival] = these_vals[ival][np.newaxis]
    these_vals = np.concatenate(these_vals,axis=0)
    all_etas[iiwt] = these_vals.reshape((36,8))
    
    these_vals = [residuals[ifit][iwt]]
    for ival in range(len(these_vals)):
        if len(these_vals[ival].shape)==1:
            these_vals[ival] = these_vals[ival][np.newaxis]
    these_vals = np.concatenate(these_vals,axis=0)
    all_residuals[iiwt] = these_vals.reshape((36,8))

model_params,model_XXs,model_YYs,model_etas,model_residuals = [x.reshape((nwt,-1)) for x in [all_params,all_XXs,all_YYs,all_etas,all_residuals]]
```
