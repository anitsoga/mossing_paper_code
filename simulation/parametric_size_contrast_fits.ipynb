```python
import nub_utils
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import scipy.optimize as sop
```

```python
RD = 77
RS = 133
R0 = 11
delta = 0
sigma_D = 7.3
sigma_S = 36.4
n = 1
m = 1.2
S,D = [lambda d: nub_utils.compute_overlap(np.array((delta,0))/d,sigma/d**2) for sigma in [sigma_S,sigma_D]]
def R(c,d):
    return R0 + RD*c**n*D(d)**m/(1+RS*c**n*S(d)**m)
```

```python
plt.figure()
ucontrast = np.array((0,6,12,25,50,100))/100
usize = np.array((5,8,13,22,36,60))
for this_size in usize:
    plt.plot(ucontrast,R(ucontrast,this_size))
```
