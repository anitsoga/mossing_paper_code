```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
```

```python
u = np.linspace(0,2,1000)
x = np.linspace(0,2,1000)
```

```python
uu,xx = np.meshgrid(u,x,indexing='ij')
```

```python
F = (1+uu)*xx/(1+uu*xx)
```

```python
plt.figure()
plt.imshow(F)
```

```python
plt.figure()
for i in range(0,1000,100):
    plt.plot(F[:,i])
plt.figure()
for i in range(0,1000,100):
    plt.plot(F[i])
```
